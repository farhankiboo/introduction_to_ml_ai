## **Review**
- Kalau mencari peluang suatu kejadian terjadi,
- **jika diketahui** kejadian lain **telah terjadi sebelumnya**,
- hal ini dinamakan dengan **CONDITIONAL PROBABILITY**

**Contoh:**
> Berapa peluang orang menggunakan produk go-ride, jika diketahui hari itu terjadi hujan?

- Peluang yang dicari adalah **orang menggunakan produk go-ride**
- Kejadian yang telah terjadi adalah **hari itu terjadi hujan**

**Untuk menghitungnya**

- Untuk menghitung *conditional probability*, dapat digunakan formula berikut.

$$
\begin{align*}
P(A \ | \ B) &= \cfrac{n(A \cap B)}{n(B)} \\
 &= \cfrac{P(A \cap B)}{P(B)}
\end{align*}
$$

- Dengan
  - $A$ adalah event yang ingin dicari peluangnya. <br>
  Contoh $A := \{ \text{orang yang menggunakan produk go-ride} \}$
  
  - $B$ adalah event yang telah terjadi. <br>
  Contoh $B := \{ \text{hari itu terjadi hujan} \}$

## **Case 2: Flu Season**

- Mendekati musim penghujan, mungkin banyak orang yang mulai terjangkit flu.
- Setiap orang akan memiliki probability untuk mendapatkan flu, $P(\text{flu})$
- Apabila seseorang mendapatkan **vaksinasi**, maka peluang mereka untuk terjangkit flu **harusnya** berubah.

- Diberikan data berikut

|Kondisi|Vaksin|Tidak Vaksin
|---|---|---|
|Flu|10|1000
|Tidak Flu|3000|10000|

---
### **Pertanyaan 1**: Dari data di atas, berapa peluang seseorang akan mengidap flu apabila telah vaksin?

---
**Jawaban**

- Kita tau yang ditanya adalah $P(\text{flu} | \text{vaksin})$
- Maka, kita bisa menjawab ini dengan conditional probability

$$
\begin{align*}
P(\text{flu} \ | \ \text{vaksin}) &= \cfrac{n(\text{flu} \cap \text{vaksin})}{n(\text{vaksin})}
\end{align*}
$$

- Langsung saja kita cari jawabannya dengan Python

````python
# Dari data
n_flu_dan_vaksin = 10
n_flu_dan_tidak_vaksin = 1000

n_tidak_flu_dan_vaksin = 3000
n_tidak_flu_dan_tidak_vaksin = 10000
````

- Dari data, $n(\text{flu} \cap \text{vaksin})=10$
- Yang tidak diketahui adalah $n(\text{vaksin})$

````python
# Kita tau bahwa jumlah orang yang vaksin adalah
n_vaksin = n_flu_dan_vaksin + n_tidak_flu_dan_vaksin

print(f"Jumlah orang yang divaksin : {n_vaksin}")
````

- Maka $P(\text{flu} | \text{vaksin})$ adalah

````python
p_flu_jika_vaksin = n_flu_dan_vaksin / n_vaksin

print(f"Peluang orang mengalami flu jika sudah vaksin : {p_flu_jika_vaksin:.4f}")
````

- **Simpulannya**, orang yang sudah vaksin ternyata memiliki peluang yang sangat kecil untuk mengidap flu

### **Pertanyaan 2**: Berapa peluang **orang menggunakan produk go-ride** jika diketahui **terjadi hujan**

- Anggap $A =: \{ \text{orang menggunakan produk go-ride} \}$
- Anggap $B =: \{ \text{terjadi hujan} \}$

````python
# LANGKAH 1/3
# Cari ruang sampel
# Jangan lupa, kejadian hujan adalah ruang sampel baru kita
# karena telah terjadi sebelumnya
n_B = 9 + 40

print(f"jumlah ruang sampel (kondisi terjadi hujan) : {n_B}")
````

````python
# LANGKAH 2/3
# Cari kejadian orang menggunakan produk go-ride (event A)
# pada kondisi hujan (event B)
# Dengan kata lain mencari A irisan B
n_A_B = 9

print(f"jumlah terjadi-nya event A dan B: {n_A_B}")
````

> **_Note_** untuk mendapatkan irisan dari 2 event, gunakan sintaks `&`. <br>
> Contoh: `data_operasi[event_A & event_B]` dibaca
> Mencari dari `data_operasi` yang memiliki kejadian `A` dan `B`

````python
# LANGKAH 3/3
# Cari Peluangnya dengan membagi n_A_B dengan n_B
peluang_A_diketahui_B = n_A_B / n_B

print(f"peluang orang menggunakan produk go-ride jika diketahui terjadi hujan adalah : {peluang_A_diketahui_B:.2f}")
````

Dapat disimpulkan bahwa peluang orang menggunakan produk go-ride jika diketahui terjadi hujan adalah 0.18

$$
P(\text{orang pakai go-ride} \ | \ \text{terjadi hujan}) = 0.18
$$

Ingat kalau

$$
P(\text{orang pakai go-ride}) = 0.43
$$

Maka
- Kejadian hujan mempengaruhi peluang orang memakai produk go-ride
- Peluang orang memakai produk go-ride menurun saat hujan terjadi
